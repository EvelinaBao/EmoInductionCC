# Examing the moderating effect of pre-existing climate belief

```python
#recode the condition to original condition coding, with control as the reference group.
long_df['Condition'] = long_df['Condition'].map({1:3, 2:2, 3:4, 4:1})
```

```python
long_df.head()
```

```python
# convert wide to long format adding pre-exisiting belief
long_df = df.melt(id_vars=['ResponseId', 'Condition', 'believer_70'],
                  value_vars=['Policy_choice_1', 'Policy_choice_2', 'Policy_choice_3', 'Policy_choice_4', 'Policy_choice_5'],
                  var_name='Policy_item',
                  value_name='Policy_choice')

long_df.head()
```

```python
%Rpush long_df
```

```python
%%R

# check is pre-believe moderate the effect of Condition
model <- glmer(Policy_choice ~ C(Condition) * C(believer_70) + (1 | ResponseId) + (1 | Policy_item),
               data = long_df, family = binomial)

print(summary(model))
summary_model <- summary(model)

cohens_d <- lme.dscore(model, data = long_df, type = "lme4")
print(cohens_d)

confint_model <- confint(model, method = "Wald")
print(confint_model)
```

# Examing the moderating effect of political affiliation

```python
# convert wide to long format adding political affiliation
long_df = df.melt(id_vars=['ResponseId', 'Condition', 'Politics'],
                  value_vars=['Policy_choice_1', 'Policy_choice_2', 'Policy_choice_3', 'Policy_choice_4', 'Policy_choice_5'],
                  var_name='Policy_item',
                  value_name='Policy_choice')

long_df.head()
```

```python
%Rpush long_df
```

```python
%%R

# check is pre-believe moderate the effect of Condition
model <- glmer(Policy_choice ~ C(Condition) * C(Politics) + (1 | ResponseId) + (1 | Policy_item),
               data = long_df, family = binomial)

print(summary(model))
summary_model <- summary(model)

cohens_d <- lme.dscore(model, data = long_df, type = "lme4")
print(cohens_d)

confint_model <- confint(model, method = "Wald")
print(confint_model)
```
